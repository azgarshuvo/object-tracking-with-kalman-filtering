```python
import pandas as pd
```

```python
my_data = pd.read_csv('pixel_pos_group_pix2.csv', index_col=None,header=None) 
my_data
```

```python
my_data_T = my_data.T 
cols = {0:'a',1:'b',2:'c',3:'d',4:'e'} 
my_data_Tc = my_data_T.rename(columns=cols)
my_data_Ts = my_data_Tc.sort_values(by=['a', 'b'], ascending =True) 
my_data_Ts = my_data_Ts.drop_duplicates(subset=['a', 'b'], keep='last')
my_data_Ts = my_data_Ts.T 
my_data_Ts 
```

```python
my_data_Ts .to_csv('pixel_pos_group_pix.csv', index=None, columns=None, header=None)  
```
